```python
import numpy as np
import re
```

```python
filepath = "/kaggle/input/multihitmon/Vocabulary.txt"
```

```python
with open(filepath, 'r') as file:
    text = file.read()

data = re.findall(r'\b[a-zA-Z]+\b', text)
data = [word.lower() for word in data]

words = list(set(data))

word_to_index = {word: idx for idx, word in enumerate(words)}

print(len(words))
```

```python
n = len(words)

word_counts = [data.count(word) for word in words]

C = np.zeros((n,n))
```

```python
k=4

for i in range(n):
    current_word = data[i]
    if current_word not in word_to_index:
        continue
    current_index = word_to_index[current_word]
    
    # Define the window bounds
    start = max(i - k, 0)
    end = min(i + k + 1, n)
    
    # Update co-occurrence counts for words within the window
    for j in range(start, end):
        if i == j:
            continue  # Skip self-co-occurrence
        neighbor_word = data[j]
        if neighbor_word in word_to_index:
            neighbor_index = word_to_index[neighbor_word]
            C[current_index, neighbor_index] += 1

# Output the co-occurrence matrix
print("Co-Occurrence Matrix:")
print(C)
```

```python
C.trace()
```

```python
for i in range(n):
    for j in range(n):
        val = min(C[i][j], 50)
        p_i_given_j = (val*n)/(word_counts[i]*word_counts[j])
        if p_i_given_j<1:
            PPMI = 0
        else:
            PPMI = np.log(p_i_given_j)
        
        C[i][j] = PPMI

print(C)
```

```python
C.trace()
```

```python
U, S, Vt = np.linalg.svd(C)

print(f"{U.shape}, {S.shape}, {Vt.shape}")

S_full = np.zeros((U.shape[0], Vt.shape[0]))
np.fill_diagonal(S_full, S)

```

```python
k = 3

U_k = U[:, :k]
S_k = S_full[:k, :k]
Vt_k = Vt[:k, :]
```

```python
A = U_k @ S_k @ Vt_k
print(A)
```

```python
W = U_k @ S_k
print(W)
print(W.shape)
```
